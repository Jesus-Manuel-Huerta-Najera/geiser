**Importar librerías**

```python
import pandas as pd
import numpy as np
import matplotlib
import sklearn
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
```

**Versiones**

```python
print("Pandas {}".format(pd.__version__))
print("Tensorflow {}".format(tf.__version__))
print("matplotlib {}".format(matplotlib.__version__))
print("sklearn {}".format(tf.__version__))
```

**Cargar el dataset**

```python
df = pd.read_csv("geisers.csv")
```

**Información del Data Frame**

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

**Graficar el DataFrame**

```python
plt.scatter(df['TimeEruption'],df['TimeWaiting'])
plt.show()
```

**Datos aleatorios**

```python
TimeEruption = np.random.randint(100,size=100)
TimeWaiting = np.random.randint(100,size=100)
 
data = {
  'TimeEruption': TimeEruption,
  'TimeWaiting':  TimeWaiting
}
df_s = pd.DataFrame(data)
 
plt.scatter(df_s['TimeEruption'],df_s['TimeWaiting'])
plt.show()
```

**Configurar Modelo**

```python
model = KMeans(n_clusters = 2,max_iter=100 )
```

**Estandarizar los datos**

```python
df = StandardScaler().fit_transform(df)
```

**Graficar valores estandarizados**

```python
plt.scatter(df[:,0],df[:,1])
plt.show()
```

**Entrenamiento**

```python
model.fit(df)
```

**Centroides**

```python
centroids = model.cluster_centers_
centroids 
```

**Ver las etiquetas**

```python
model.labels_
```

**Graficar los centroides**

```python
plt.figure(figsize=(8,8))
plt.scatter(df[:,0],df[:,1],c=model.labels_.astype(float),s=10,label="clusters")
plt.scatter(centroids[:,0],centroids[:,1],c="red",marker="*",s=20,label="centroids")
plt.show()
```

**Predicciones**

```python
new_label = model.predict([[-1.0,0.0]])
new_label[0]
```

**Seleccionar el número clusters**

```python
ess =  [] #  Suma de las distancia al cuadrado.
lista_k = list(range(1,10))

for k in lista_k:
	model = KMeans(n_clusters=k)
	model.fit(df_s)
	ess.append(model.inertia_)

plt.figure(figsize=(8,8))
plt.plot(lista_k, ess, "-*")
plt.xlabel("Clusters")
plt.ylabel("ess")

plt.show()
```

**Graficar con distintos clusters**

```python
lista_k = list(range(1,10))
 
for k in lista_k:
  model = KMeans(n_clusters=k)
  model.fit(df)
  ess.append(model.inertia_)
  centroids = model.cluster_centers_
  plt.figure(figsize=(8,8))
 
  plt.scatter(df[:,0],df[:,1],c=model.labels_.astype(float),s=10,label="clusters")
  plt.scatter(centroids[:,0],centroids[:,1],c="red",marker="*",s=20,label="centroids")
 
  plt.show()
```
